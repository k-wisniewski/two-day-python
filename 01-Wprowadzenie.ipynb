# Czym jest Python i do czego można go zastosować?

Python to dynamicznie typowany, interpretowany język ogólnego przeznaczenia. Robi się w nim właściwie wszystko, bo do większości zastosowań jest po prostu wystarczająco dobry. Jego bolączką bywa wydajność i trudna relacja z współbieżnością, ale zazwyczaj albo to nie jest problem, albo jest już moduł napisany w C, który to obchodzi.

## Język kompilowany czy interpretowany?
Mimo że Python jest uważany za język interpretowany, w rzeczywistości sprawa jest bardziej skomplikowana. Po pierwsze istnieje wiele implementacji języków programowania, w tym Pythona (CPython jest "kanoniczną", ale jest też wiele innych np. pypy, Cython, IronPython etc.), więc nie można powiedzieć o języku samym w sobie, że jest "interpretowany". Jest natomiast prawdą, że zdecydowana większość użytkowników nigdy nie używało innych niż CPython implementacji, więc dla nich te rozważania mogą wydawać się teoretyzowaniem.

Implementacje języków programowania dzielą się na dwie główne klasy: kompilowane i interpretowane. Implementacje kompilowane wymagają budowania, czyli tłumaczenia kodu napisanego przez programistę na kod maszynowy, wykonywalny przez komputer. Bez tego kod jest dla komputera kompletnie niezrozumiały. Plusem jest natomiast możliwość optymalizacji kodu przed wykonaniem i możliwość wyprodukowania samowystarczalnego pliku wykonywalnego.

Języki interpretowane umożliwiają pisanie programów, które można od razu uruchomić - ich kod jest - nomen omen -interpretowany, lub mówiąc prościej wykonywany przez interpreter. W najprostszej formie kod napisany przez programistę jest wprost "rozumiany" przez program interpretera, jednak ze względów wydajnościowych jest to rozwiązanie niepraktyczne i rzadko stosowane. Zazwyczaj kod jest przerabiany na prostszą reprezentację pośrednią (IR), optymalizowany, a interpreter wykonuje już wstępnie przetworzony kod. W przypadku najpopularniejszej, "kanonicznej" implementacji Pythona mamy do czynienia właśnie z taką sytuacją - polecenie `python` w rzeczywistości kompiluje kod do tzw. bytecode'u, cache'owanego w postaci plików `.pyc`, a następnie szybka maszyna wirtualna wykonuje bytecode. Dzięki temu zyskujemy przenośność kodu charakterystyczną dla implementacji interpretowanych oraz akceptowalną do większości zastosowań szybkość. Ceną jest konieczność zainstalowania interpretera Pythona - jeśli jest to problem, warto rozważyć implementacje takie jak Cython, które kompilują podzbiór Pythona do kodu maszynowego.
## Paradygmaty programowania obsługiwane przez Pythona
Python jest językiem wieloparadygmatowym, jednak szczególny nacisk położony jest na implementacje nastepujących paradygmatów:
1. Proceduralnego
3. Obiektowego
4. Funkcyjnego

### Python jako język proceduralny
Podstawowym mechanizmem abstrakcji w Pythonie jest funkcja. Kod można dzielić na funkcje, które następnie można wywoływać z parametrami, zwiększając reużywalność określonych fragmentów. Wewnątrz funkcji zdefiniowane są kolejno wykonywane sekwencje instrukcji opisujące jak otrzymać porządany wynik. Składnia Pythona wywodzi się z języków ABC, Algol 68 czy Modula

### Python jako język obiektowy
Paradygmat obbiektowy również jest wspierany - możliwe jest definiowanie klas enkapsulujących dane i zachowania w sposób podobny do innych języków obiektowych. Wiele wewnętrznych mechanizmów Pythona jest zresztą zaimplementowana przy użyciu obiektowości - np. kolekcje są obiektami implementującymi interfejsy, które umożliwiają odwoływanie się do konkretnych ich elementów, iterowanie itp. Nawet funkcje w Pythonie są po prostu obiektami, implementującymi interfejs `Callable`! Na pierwszy rzut oka obiektowość w Pythonie jest zaimplementowana w podobnym stylu co w C++ czy Javie - obiekty należą do klas, posiadają swoje pola i implementują swoje zachowania w postaci metod. W przeciwieństwie jednak do bardziej tradycyjnych języków obiektowych - np. Java lub C#, nie są dostępne niezawodne mechanizmy ograniczania widoczności atrybutów obiektów, co zostanie omówione w jednym z kolejnych modułów szkolenia. Tak samo inaczej niż w Javie jest możliwe dynamiczne dodawanie atrybutów i zachowań do już istniejących egzemplarzy klas, również wówczas gdy inne egzemplarze pozostają niezmienione.

### Python jako język funkcyjny
Funkcje w Pythonie są "obywatelami pierwszej kategorii". Można je przekazywać jako parametry do innych funkcji, przypisywać do zmiennych czy zwracać jako wyniki wywołania funkcji. W bibliotece standardowej znajdują się znane z innych funkcyjnych języków funkcje takie jak `map`, `filter`, `reduce`, możliwe jest zagnieżdżanie i tworzenie domknięć, a nawet tworzenie całych programów w stylu funkcyjnym. Składnia natomiast umożliwia bardzo zwięzłe przekształcenia kolekcji czy potoków przy użyciu wyrażeń "comprehensions". Nie jest to jednak typowe zastosowanie Pythona, oficjalne przewodniki opisujące styl kodowania w Pythonie raczej tego nie zalecają a domyślna implementacja - CPython nie ma nawet podstawowych optymalizacji np. rekursji ogonowej.

Oprócz tych 3 paradygmatów wspierane jest też programowanie zorientowane aspektowo (przez dekoratory), metaprogramowanie (przez metaklasy), a nawet istnieją biblioteki umożliwiające programowanie w logice a'la Prolog. Zdecydowana większość programistów skupia jednak swoje wysiłki na zastosowaniach omówionych tu 3 paradygmatów, rzadko mając głębszą styczność z pozostałymi (poza użyciami już istniejących frameworków czy bibliotek).

### Process PEP
Python Enhancement Proposals, znane jako PEP, są fundamentalnym elementem rozwoju języka Python. Są to szczegółowe dokumenty techniczne, które przedstawiają nowe funkcjonalności, zmiany w języku lub jego bibliotekach, a także różne standardy i konwencje. Proces PEP jest kluczowy dla ewolucji Pythona, umożliwiając zarówno twórcom języka, jak i jego użytkownikom kształtowanie jego przyszłości. Każdy PEP przechodzi przez kilka etapów: od draftu, przez dyskusję, do ostatecznej akceptacji lub odrzucenia.

PEP-y często dotyczą ważnych aspektów języka, jak formatowanie kodu (PEP 8), czy filozofii języka (PEP 20 - 'Zen Pythona'). Istnieją jednak również PEP-y o charakterze bardziej technicznym, mające znaczący wpływ na praktyki programistyczne. Przykładem może być PEP 517, który wprowadza nowe standardy dla systemu budowania i dystrybucji pakietów w Pythonie. PEP ten proponuje oddzielenie procesu budowania pakietów od ich instalacji, co ma na celu uproszczenie i zwiększenie elastyczności tego procesu. PEP 517 jest odpowiedzią na rosnące wymagania i złożoność systemów zarządzania pakietami w Pythonie, oferując bardziej modularne i elastyczne podejście.

Decyzje dotyczące przyjęcia lub odrzucenia Python Enhancement Proposal (PEP) są podejmowane przez tzw. Python Steering Council, który zastąpił wcześniej funkcjonującą rolę BDFL (Benevolent Dictator For Life) - rolę tę pełnił twórca Pythona, Guido van Rossum, do lipca 2018 roku. Steering Council jest wybierany przez społeczność Pythona i składa się z pięciu członków, którzy mają szerokie doświadczenie w pracy nad Pythonem. Ich zadaniem jest kierowanie rozwojem języka i podejmowanie decyzji o przyszłości Pythona, w tym o przyjmowaniu lub odrzucaniu PEP-ów.

## *Konfiguracja środowiska*
Aby skonfigurować lokalne środowisko potrzebne nam do pracy należy spełnić wymagania systemowe z sekcji poniżej i sklonować sobie repozytorium poleceniem:
```
git clone git@github.com:k-wisniewski/two-day-python.git
```
Należy stworzyć nowe środowisko wirtualne:
```
python -m venv env
```
a następnie zainstalować tam zależności poleceniem:
```
python -m pip install -r requirements.txt
```

Powyższe instrukcje zakładają, że `which python` zwraca właściwą instalację interpretera Pythona w wersji 3.12. W przeciwnym przypadku - należy dokładnie wyspecyfikować o którą instalacje nam chodzi.
